```python
import agentpy as ap
from owlready2 import *
from matplotlib import pyplot as plt
import IPython
```

```python
class VaccumAgent(ap.Agent):
    def setup(self):
        self.agentType = 1
        pass

    def step(self):
        self.forward()
        pass

    def update(self):
        pass

    def end(self):
        pass

    #ACCIONES
    def forward(self):
        self.model.grid.move_by(self, (1,1))
```

```python
class DirtAgent(ap.Agent):
    def setup(self):
        self.agentType = 2
        pass

    def step(self):
        pass

    def update(self):
        pass

    def end(self):
        pass
```

```python
class VaccumModel(ap.Model):
    def setup(self):
        self.grid = ap.Grid(self,(self.p.M,self.p.N), track_empty=True)
        self.agents = ap.AgentList(self, self.p.vaccums, VaccumAgent)
        self.dirtList = ap.AgentList(self, self.p.dirty_places, DirtAgent)
        self.grid.add_agents(self.agents, random=True)
        self.grid.add_agents(self.dirtList, random=True)
        pass

    def step(self):
        self.agents.step()
        pass

    def end(self):
        pass

    def update(self):
        pass

    def end(self):
        pass
```

```python
parameters={
    'vaccums':1,
    'dirty_places': 10,
    'M' : 10,
    'N' : 10,
    'steps' :100,
    'seed' : 12345,
}

#model = VaccumModel(parameters)
#results = model.run()
```

```python
#A FUNCTION TO ANIMATE THEE SIMULATION

def animation_plot(model, ax):
    agent_type_grid = model.grid.attr_grid('agentType')
    ap.gridplot(agent_type_grid, cmap='Accent', ax=ax)
    ax.set_title(f"Vaccum Model \n Time-step: {model.t}, "
                 f"dirty: {0}")
```

```python
#SIMULATION:

#Create figure (from matplotlib)
fig, ax = plt.subplots()

#Create model
model = VaccumModel(parameters)

#Run with animation
#If you want to run it without animation then use instead:
#model.run()
animation = ap.animate(model, fig, ax, animation_plot)
#This step may take a while before you can see anything

#Print the final animation
IPython.display.HTML(animation.to_jshtml())
```
